## Connection with data and checking it

```python
df = pd.read_csv('courses.csv')
```

```python
df.head()
```

```python
df.tail()
```

```python
## Data prepare

categories = df['category'].value_counts()
```

```python
categories.plot(kind='barh', color='blue' ,figsize=(10,8))
plt.xlabel('Number of Courses')
plt.ylabel('Category')
plt.title('Number of Courses per Category')
plt.show()
```

```python
languages = df['language'].value_counts()
```

```python
languages.plot(kind='barh', color='blue' ,figsize=(10,8))
plt.xlabel('Number of Courses')
plt.ylabel('Category')
plt.title('Number of Courses per Category')
plt.show()
```
